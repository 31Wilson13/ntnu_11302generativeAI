```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 畫函數圖形
\begin{equation}
y = e^{-\alpha x} \cdot (\sin(\omega x) + A\sin(n\omega x)) \cdot (1 + \beta x)
\end{equation}
- 這個函數模擬了能量的衰減，振幅隨著時間的前進，也就是$X$軸的增加，振幅逐漸縮小，最後歸零
- 以彩虹色彩的轉變搭配顯示振幅的衰減
- 可以透過調整程式碼參數來改變函數圖形的顯示：
    1. 衰減係數 α：控制振幅減小的速度
    2. 基本頻率 ω：改變基本振動週期
    3. 諧波係數 n：調整附加振動的頻率
    4. 諧波振幅 A：控制附加振動的強度
    5. 線性調整係數 β：改變振幅的緩慢變化率

```python
from matplotlib.collections import LineCollection

alpha = 0.2    # 衰減係數
omega = 2      # 基本頻率
A = 0.5        # 諧波振幅
n = 1          # 諧波頻率倍數
beta = 0.1     # 線性調整係數


x = np.linspace(0, 20, 2000)
y = np.exp(-alpha*x) * (np.sin(omega*x) + A*np.sin(n*omega*x)) * (1 + beta*x)


points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
norm = plt.Normalize(0, len(segments))
lc = LineCollection(segments, cmap='rainbow', norm=norm)
lc.set_array(np.linspace(0, len(segments), len(segments)))

fig, ax = plt.subplots(figsize=(12, 6))
ax.add_collection(lc)
ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())
plt.title('Damped Oscillation')
plt.grid(True)
```

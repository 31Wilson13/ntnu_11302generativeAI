# 載入必要套件

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 畫函數圖形
\begin{equation}
y = e^{-\alpha x} \cdot (\sin(\omega x) + A\sin(n\omega x)) \cdot (1 + \beta x)
\end{equation}
- 這個圖形模擬了能量的衰減情形，振幅隨著時間的前進，也就是$X$軸的增加，振幅逐漸縮小，最後漸漸歸0
- 以彩虹色彩的轉變搭配顯示振幅的衰減
- 可以透過調整程式碼參數來改變函數圖形的顯示：
    1. 衰減係數 α：控制振幅減小的速度
    2. 基本頻率 ω：改變基本振動週期
    3. 諧波係數 n：調整附加振動的頻率
    4. 諧波振幅 A：控制附加振動的強度
    5. 線性調整係數 β：改變振幅的緩慢變化率

```python
from matplotlib.collections import LineCollection

# 參數設置
alpha, omega, A, n, beta = 0.2, 2, 0.5, 1, 0.1
x_range = (0, 25)
points_count = 1000

x = np.linspace(*x_range, points_count)

# 函數方程式
exp_term = np.exp(-alpha*x)
sin_term = np.sin(omega*x) + A*np.sin(n*omega*x)
linear_term = 1 + beta*x
y = exp_term * sin_term * linear_term

# 將函數圖形切分為無數個線段，以利後續繪製彩色圖形
points = np.column_stack([x, y]).reshape(-1, 1, 2)
segments = np.hstack([points[:-1], points[1:]])

# 繪圖程式碼
fig, ax = plt.subplots(figsize=(12, 6))
lc = LineCollection(segments, cmap='rainbow') #產生彩色線條
lc.set_array(np.linspace(0, 1, len(segments)))

ax.add_collection(lc)
ax.set_xlim(x_range) #設置X軸範圍
ax.set_ylim(np.min(y), np.max(y)) #設置Y軸範圍
ax.set_title('Damped Oscillation')
ax.grid(True) #顯示圖形內格線
```
